# only compare the genes which are among the top 623 genes(up and down regulated genes in breast cancer)

```python
# read top up and down regulated genes data
top_genes_data = pd.read_excel('/Users/yanlixu/Desktop/pingan/合作/CDK12/220524/5_compounds_for_drug_combination.xlsx', sheet_name='top_623_genes')

top_genes_data = top_genes_data.set_index('Name_GeneSymbol')
top_genes_data
```

```python
# merge two dataframe
df_merged = pd.merge(diff_value_df, top_genes_data[['Value_LogDiffExp']], left_index=True, right_index=True)
df_merged.rename(columns={'Value_LogDiffExp': 'breast_cancer_reverse_gene_expression'}, inplace=True)
df_merged
```

# Calculation 1: compounds compare with 'LINCSCP_32475'

```python
# if the expression value same direction (both upregulated or both down regulated) with LINCSCP_32475, then 1; else 0
def compare(col1, col2):
    result = col1 * col2
    return [1 if res > 0 else 0 for res in result] 

df_merged_1 = df_merged.drop(columns=['breast_cancer_reverse_gene_expression'])
result_1 = []
for col in df_merged_1.columns.tolist():
    result_1.append(compare(df_merged_1['LINCSCP_32475'], df_merged_1[col]))
result_1_df = pd.DataFrame(result_1, columns = df_merged_1.index.tolist())
result_1_df.index = df_merged_1.columns
result_1_df
```

```python
ax = sns.clustermap(result_1_df, standard_scale=1, cmap="vlag")
plt.show()
```

```python
# how many genes same direction(1s) with LINCSCP_32475

result_1_df['sum_of_same_dir_genes'] = result_1_df.sum(axis=1)
result_1_df['sum_of_same_dir_genes_rate'] = round(result_1_df['sum_of_same_dir_genes']/41, 2)
result_1_df
```

# Calculation 2: compounds compare with reverse_of_breast_cancer

```python
# if the expression value same direction (both upregulated or both down regulated) with reverse_of_breast_cancer, then 1; else 0


def compare(col1, col2):
    result = col1 * col2
    return [1 if res > 0 else 0 for res in result] 

#df_merged_1 = df_merged.drop(columns=['breast_cancer_reverse_gene_expression'])
result_2 = []
for col in df_merged.columns.tolist():
    result_2.append(compare(df_merged['breast_cancer_reverse_gene_expression'], df_merged[col]))
result_2_df = pd.DataFrame(result_2, columns = df_merged.index.tolist())
result_2_df.index = df_merged.columns
result_2_df
```

```python
ax = sns.clustermap(result_2_df, standard_scale=1, cmap="vlag")
plt.show()
```

```python
# how many same direction with reverse_of_breast_cancer

result_2_df['sum_of_same_dir_genes'] = result_2_df.sum(axis=1)
result_2_df['sum_of_same_dir_genes_rate'] = round(result_2_df['sum_of_same_dir_genes']/41, 2)
result_2_df
```

# merge two dataframe

```python
df_final = pd.merge(result_1_df[['sum_of_same_dir_genes_rate']], result_2_df[['sum_of_same_dir_genes_rate']], left_index=True, right_index=True)
df_final.rename(columns={'sum_of_same_dir_genes_rate_x': 'percentage_of_same_direction_genes_w_32475', 'sum_of_same_dir_genes_rate_y': 'percentage_of_reverse_direction_genes_w_breast_cancer'}, inplace=True)

df_final['rank_score'] = round(df_final['percentage_of_reverse_direction_genes_w_breast_cancer']/df_final['percentage_of_same_direction_genes_w_32475'], 2)

```

```python
df_final = df_final.sort_values(by=['rank_score'], ascending=False)
df_final
```

```python
sns.scatterplot(data=df_final, x="percentage_of_same_direction_genes_w_32475", y="percentage_of_reverse_direction_genes_w_breast_cancer")
plt.show()


```
