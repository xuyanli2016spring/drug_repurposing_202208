# get all the her2_breast cancer drugs

```python
drug_target_file = '/Users/yanlixu/Desktop/pingan/合作/CDK12/network_based/1_her2_breast_cancer_20220620.xlsx'
drug_targets = pd.read_excel(drug_target_file, sheet_name='5_drug_target_w_diamond_100')
drug_list = drug_targets['DrugID (DrugBank ID)'].tolist()
drug_list_unique =list(set(drug_list))
drug_list_unique
```

```python
df_all_combs = pd.read_csv('all_combs_unormalized.csv')
df_all_combs
```

```python
# get all the breast cancer drug combinations

drug_comb_res = []
for drug in drug_list_unique:
    print(drug)
    for index, row in df_all_combs.iterrows():
        if drug in row['drugbank_identifiers']:
            drug_comb_res.append([drug, row['index'], row['drugs'], row['drugbank_identifiers'], row['pubchem_identifiers'], row['source_id'], row['source']])
            
df_drug_comb_res = pd.DataFrame(drug_comb_res, columns = ['her2_drug', 'index', 'drugs', 'drugbank_identifiers', 'pubchem_identifiers', 'source_id', 'source'])
df_drug_comb_res 
    
    
```

```python
df_drug_comb_res.to_csv('her2_combs.csv')
```
